## Parameter 

```python
## Parameter 
start_date = date(2023, 10, 16)
end_date = date(2023, 10, 20)
```

```python
# Initialize an empty DataFrame
df_postorderstatus_requesting = pd.DataFrame()

# Loop through the date range day by day using a for loop
for current_date in (start_date + timedelta(n) for n in range((end_date - start_date).days + 1)):
    
    current_date = current_date.strftime('%Y%m%d')
    # Construct the query with the current date
    query = f"""
        SELECT 
            *,
            SUBSTR(profile_phone, 3) customer_mobile,
            json_extract_scalar(eventProps_data,'$.orderId') order_id
        FROM
            raw.clevertap_customer_postorderstatus_requesting
        WHERE 
            yyyymmdd = '{current_date}'
    """
    
    # Execute the query and append the results to the DataFrame
    daily_data = pd.read_sql(query, connection)
    df_postorderstatus_requesting = df_postorderstatus_requesting.append(daily_data, ignore_index=True)

# Print the first few rows of the combined DataFrame
df_postorderstatus_requesting
```

```python
df_postorderstatus_requesting.to_csv('raw_postorderstatus_requesting_{}_to_{}.csv'.format(start_date, end_date), 
                                     index=False)
```
