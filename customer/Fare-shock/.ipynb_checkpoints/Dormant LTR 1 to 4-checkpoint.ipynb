```python
## DORMANT_LTR_1to4

dormant_ltr = f"""
        with PAN as (
            select 
                 *
            from
            (
                select
                    customer_id,
                    run_date,
                    case 
                        when taxi_recency_segment in ('DORMANT') and taxi_lifetime_rides IN (1,2,3,4) then 'DORMANT_LTR_1to4'
                        when taxi_recency_segment in ('RECENT','STATIONARY') and taxi_lifetime_rides IN (1,2,3,4) then 'ACTIVE_LTR_1to4'
                        when taxi_recency_segment in ('RECENT','STATIONARY') and taxi_lifetime_rides > 4 then 'Active_Phh'
                        when taxi_recency_segment in ('DORMANT') and taxi_lifetime_rides > 4 then 'DORMANT_Phh'
                        when taxi_recency_segment in ('INACTIVE') then taxi_recency_segment
                        when taxi_lifetime_rides = 0 then 'Allocator_Ltr_0' 
                        end ltr_bucket
                from 
                    datasets.iallocator_customer_segments 
                where 
                    run_date = date_format(DATE('{start_date}') - interval '1' day,'%Y-%m-%d')
                    and taxi_recency_segment in ('RECENT','STATIONARY','DORMANT')
                )
                where ltr_bucket in ('Active_Phh', 'DORMANT_LTR_1to4')
            ),

            registered2iall as (

            select
                customerId,
                registrationDate,
                case 
                    when registrationDate <= DATE('{start_date}') then 'registered_before_run_date'
                    when registrationDate > DATE('{start_date}') and registrationDate <= DATE('{start_date}') + interval '30' day then 'registered_in_performance_month'
                    else 'other' 
                    end registration_customer_type,
                a.* 
            from 
                PAN a 
            right join 
                datasets_internal.customer_single_view_v5 b 
                on a.customer_id = b.customerId 

            where 
                registrationDate <= DATE('{start_date}') + interval '30' day
            ),

            funnel as (

            select
                customerid,
                day,
                ao_sessions_unique_daily as ao,
                fe_sessions_unique_daily as fe,
                rr_sessions_unique_daily as rr,
                net_rides_daily as net_rides,
                gross_rides_daily as gross_rides

            from 
                datasets.customer_rf_daily_kpi

            where 
                cast(day as date)>= DATE('{start_date}')
                and cast(day as date) <= DATE('{start_date}') + interval '30' day
                and service_name in ('Auto','Link')
            ),

            ltr_segment as (

            select
                ltr_bucket_u,
                customerId
            from    
                (
                select
                    *,
                    case 
                        when ltr_bucket is not null and registration_customer_type = 'registered_in_performance_month' then null 
                        else ltr_bucket 
                    end ltr_bucket_u,
                    DATE_diff('month',registrationDate,DATE('{start_date}') - interval '1' day) registered_to_run_in_months
                from
                (
                    select
                        customerid as daily_kpi_customer,
                        -- count(distinct customerid)
                        sum(gross_rides) gross_rides,
                        sum(net_rides) net_rides,
                        sum(rr) as rr
                    from 
                        funnel
                    group by 1
                ) a 
                full join 
                    registered2iall b 
                    on a.daily_kpi_customer = b.customerId
                )
            where 
                registration_customer_type is not null 
                and ltr_bucket_u in ('Active_Phh','DORMANT_LTR_1to4')
            )
            select 
                *
            from 
                ltr_segment
"""

df_dormant_ltr = pd.read_sql(dormant_ltr, connection)
df_dormant_ltr.head()
```

```python
startdate = '20231002'
enddate = '20231008'
```

```python
## merged_tbl

merged_tbl = f"""
    with orders_tbl as 
            ( SELECT
                    orderdate,
                    city,
                    service_name,
                    estimate_id, 
                    fare_recalculated_reason reason,
                    customer_id,
                    order_id,
                    order_amount,
                    fe_re_type,
                    fe_re_diff_amount,
                    u.id_array as fare_estimate_id
                FROM
                    (
                    SELECT   
                        city_name as city,
                        service_obj_service_name as service_name,
                        order_id,
                        customer_id,
                        amount as order_amount,
                        date_format(from_unixtime(epoch / 1000, 'Asia/Kolkata'), '%Y-%m-%d') AS orderdate,
                        epoch,
                        estimate_id,
                        fare_recalculated_reason,
                        fare_recalculated_type AS fe_re_type,
                        fare_recalculated_diff_amount AS fe_re_diff_amount,
                        cast(json_parse(estimate_ids) as array<varchar>) AS id_array
                    FROM
                        orders.order_logs_snapshot
                    WHERE
                        yyyymmdd >= '{startdate}'
                        AND yyyymmdd <= '{enddate}' 
                        AND city_name is not null
                        AND order_status in ('dropped')
                        AND spd_fraud_flag != True
                        AND service_obj_service_name IN ('Link', 'Auto')
                    ) as t
                CROSS JOIN UNNEST(t.id_array) AS u(id_array)),

        fe_tbl as (
                SELECT
                    city AS city,
                    user_id as customer_id,
                    fare_estimate_id,
                    service_name,
                    api_context,
                    date_format(from_unixtime(epoch / 1000, 'Asia/Kolkata'), '%Y-%m-%d') AS orderdate,
                    cast(final_amount AS double) AS fe_amount,
                    epoch
                FROM
                    pricing.fare_estimates_enriched
                WHERE
                    yyyymmdd >= '{startdate}'
                    AND yyyymmdd <= '{enddate}'
                    AND service_name IN ('Link', 'Auto')

        ),


        merged_tbl as (
            SELECT 
                *,
                CASE 
                WHEN amt_diff < -10 THEN '< -10'
                WHEN amt_diff >= -10 AND amt_diff < -7 THEN '-10 to -7'
                WHEN amt_diff >= -7 AND amt_diff < -5 THEN '-7 to -5'
                WHEN amt_diff >= -5 AND amt_diff < 0 THEN '-5 to 0'
                WHEN amt_diff > 0 AND amt_diff <= 5 THEN '0 to 5'
                WHEN amt_diff > 5 AND amt_diff <= 7 THEN '5 to 7'
                WHEN amt_diff > 7 AND amt_diff <= 10 THEN '7 to 10'
                WHEN amt_diff > 10 THEN '> 10'
                END AS amt_diff_bucket
            FROM
            (
            SELECT  
                o.orderdate,
                o.service_name,
                o.customer_id,
                o.order_id,
                o.estimate_id,
                api_context,
                o.fare_estimate_id,
                o.reason,
                fe_amount,
                o.order_amount,
                o.order_amount -  fe_amount AS amt_diff,
                fe_re_type,
                fe_re_diff_amount,
                row_number() over (partition by o.estimate_id order by f.epoch asc)  as updated_seq 
            From orders_tbl o
            INNER JOIN fe_tbl f
            ON o.fare_estimate_id = f.fare_estimate_id
            AND o.service_name = f.service_name
            )
        )
        
        select 
            *
        from 
            merged_tbl
        where 
            order_id in (SELECT order_id from merged_tbl where  amt_diff!=0)
            and order_id not in (SELECT order_id from merged_tbl where api_context = '/fare/editOrder')
            and updated_seq = 1

"""

merged_tbl = pd.read_sql(merged_tbl, connection)
merged_tbl.head()
```

```python
df_join = pd.merge(df_dormant_ltr,
                   merged_tbl,
                   how = 'inner',
                   left_on = ['customerId'],
                   right_on = ['customer_id']
                    )
df_join.head()
```

```python
agg_result = df_join \
                .groupby(['ltr_bucket_u', 'service_name', 'reason']) \
                .agg(
#                     date_range=pd.NamedAgg(column='orderdate', aggfunc=lambda x: f"{x.min()} to {x.max()}"),
                    total_shocked_orders=pd.NamedAgg(column='order_id', aggfunc='nunique')
                ) \
                .reset_index()
agg_result = agg_result.sort_values(['service_name', 'ltr_bucket_u', 'total_shocked_orders'], ascending=False)
agg_result
# agg_result['']
```
